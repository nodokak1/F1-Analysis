## Who are the top 10 drivers of all time?

To answer this question, we can look at it from a quantitative perspective by ranking the drivers based on their career championship points, as well as total wins in their career.

### Examining tables used

```python
print('drivers')
print(pd.read_sql('SELECT * FROM drivers', conn).head())
print('driver_standings')
print(pd.read_sql('SELECT * FROM driver_standings', conn).head())
print('races')
print(pd.read_sql('SELECT * FROM races', conn).head())
```

### Career championship points ranking

An important point to take note is that `points` in the `driver_standings` table refers to the total points earned by the driver in the season, and not in the single race alone.

Thus, in addition to simply joining the `drivers` and `driver_standings` table (join on `driverId`: primary key of `drivers` table), an additional join to the `races` tables are required. The `races` table can be joined using `raceId` (primary key of `races` table).

```python
pd.read_sql("""SELECT (forename || ' '|| surname) AS driver, dob AS date_of_birth, SUM(max_points_per_year) AS career_points
FROM (
    SELECT d.driverId, d.dob, d.forename, d.surname, MAX(ds.points) AS max_points_per_year
    FROM drivers d
    LEFT JOIN driver_standings ds 
    USING (driverId)
    LEFT JOIN races r 
    USING (raceId)
    GROUP BY d.driverId, r.year
)
GROUP BY driverId
ORDER BY career_points DESC
LIMIT 10
""", conn)
```

### Total career wins

To answer this question through a driver's total career wins, we can use the same tables to derive their rankings.

In the `driver_standings` table, `wins` represents the total number of wins for the driver in that season. 

```python
pd.read_sql("""SELECT (forename || ' ' || surname) AS driver, dob AS date_of_birth, SUM(max_wins_per_year) AS total_wins
FROM (
    SELECT d.driverId, d.dob, d.forename, d.surname, MAX(ds.wins) AS max_wins_per_year
    FROM drivers d
    LEFT JOIN driver_standings ds
    USING (driverId)
    LEFT JOIN races r
    USING (raceId)
    GROUP BY d.driverId, r.year
)
GROUP BY driverId
ORDER BY total_wins DESC
LIMIT 10
""", conn)
```

### Evaluation and further analysis

We can thus see that these 2 different ways to qualitatively measure how good a driver is does derive different rankings. 

An intruiging observation is that in the ranking of drivers using total career wins, we see that there are more drivers from earlier on in Formula 1 history. For instance, Alain Prost, Ayrton Senna, Nigel Mansell, Jackie Stewart and Jim Clark. To explain this, we will conduct more analysis on the points system, as well as the number of F1 races from 1950 to present day.

#### Points system in F1 history

To analyse the points system throughout F1 history, we can use the `results` table to see the points awarded to drivers in the first race of the season (*points in this table refers to total points of a driver in a season*). 

```python
df = pd.read_sql("""SELECT r.decade, p.points, p.position
FROM (
    SELECT raceId, year, (year / 10) * 10 AS decade
    FROM races
    GROUP BY decade
    HAVING MIN(raceId)
) r
JOIN results p
USING (raceId)
WHERE p.points != 0.0
ORDER BY r.decade
""", conn)

# pivot wider such that each decade is a column, and the position is a row
pivot = df.pivot(index = "position", columns = "decade", values = "points")

# sort the positions in the correct order
pivot = pivot.reindex(sorted(pivot.index, key = int), axis=0)
pivot
```

We can observe that throughout the decades, the points system was modified such that more points are being awarded to drivers in recent years. Hence, it can explain how even though drivers in the past may have gotten more wins, they are not part of the rankings if we measure it by career championship  points. Comparing the 1950s to the 2020s, we can see that there is a 177.778% increase (9 to 25) in the number of championship points awarded to the driver who wins the race. 

To further exacerbate the issue, in the 1950s, points were only awarded to the top 5 drivers while in the 2020s, points are awarded to the top 10 drivers. 

#### Number of F1 races throughout history

We can plot a line graph to show the changes in the number of F1 races throughout time. To do this, we can utilise the `races` table.

```python
# preparing data
df = pd.read_sql("""SELECT year, COUNT(*) as num_races
FROM races
GROUP BY year
""", conn)

y_values = df["num_races"]
x_values = df["year"]

decades = df[df["year"].astype(str).str.endswith("0")]
display(decades.head())
display(df.head())
```

From the graph, we can also see that the number of races in F1 have increased over time from 1950 to 2024. This, coupled with the increase in the championship points awarded at every race, explains why there are only drivers from recent years when we rank them based on championship points.

All in all, it is difficult to measure who exactly are the top 10 drivers of all time due to many contributing factors such as the car that they are driving, and as we have seen above, changes to F1 regulations. However, a better measure of how good a driver is is to  measure their performance against their teammate's performance.

To do this, we can find the performance of the driver and their teammate at each race, and calculate the average difference between their position. The driver with the higher average difference would then be the better driver.

### Better metric to find top 10 drivers

In order to compare a driver with their teammate at each race, the `drivers`, `results` and `constructors` table will need to be utilized.

```python
df = pd.read_sql("""SELECT d.forename, d.surname, c.name AS constructor_name, r.points, r.raceId, r.constructorId, ra.year, d.driverId
FROM drivers d
LEFT JOIN results r
USING (driverId)
LEFT JOIN constructors c
USING (constructorId)
LEFT JOIN races ra
USING (raceId)
""", conn)

df = df[['raceId', 'constructorId', 'driverId', 'points', 'forename', 'surname', 'constructor_name', 'year']]

# Define a function to calculate point differences
def calculate_point_difference(group):
    group = group.sort_values(by='points', ascending=False).reset_index(drop=True)
    
    if len(group) > 1:  # Ensure there are at least two drivers in the team
        point_diff = group.iloc[0]['points'] - group.iloc[1]['points']
        return pd.DataFrame({
            'driverId': [group.iloc[0]['driverId'], group.iloc[1]['driverId']],
            'point_difference': [point_diff, -point_diff],  # + for driver1, - for driver2
            'constructor_name': group.iloc[0]['constructor_name'],
            'year': group.iloc[0]['year']
        })
    else:
        return pd.DataFrame({
            'driverId': [group.iloc[0]['driverId']],
            'point_difference': [0],  # No difference to record
            'constructor_name': [group.iloc[0]['constructor_name']],
            'year': [group.iloc[0]['year']]
        })
        
# Apply the function to each group
df_diff = df.groupby(['constructorId', 'raceId']).apply(calculate_point_difference).reset_index(drop=True)

# Sum point differences for each driver across all races
df_sum = df_diff.groupby('driverId')['point_difference'].sum().reset_index()
df_sum = df_sum.merge(df[['driverId', 'forename', 'surname']].drop_duplicates(), on='driverId').sort_values(by = 'point_difference', ascending = False)
df_sum = df_sum.drop(columns = ['driverId'])

# Display the results
print(df_sum.head(10))
```

Using the point difference of each driver with their teammate, we again arrive at a different ranking of the top 10 drivers. However, compared to the total championship points ranking, this again can be affected by the reasons stated above. 

While it is not the perfect metric to score drivers on, it reduces the amount of uncertainty in rankings that are affected by the car that the driver is driving, as well as the conditions on the track at that race. This is because we are comparing the drivers to their teammate directly, meaning that the conditions on the track, as well as the car being used are the same, and thus we are comparing the drivers purely based on their skills.

In addition, another factor that affects this ranking, as well as the 2 used initially,is the length of a driver's career. Especially in this ranking, where every point earned is used in the metric, the difference is further exacerbated. The top driver in this ranking, Fernando Alonso, has the longest F1 career in history that spans 23 years. Because of this, he would then have more opportunities compared to all the other drivers to be able to score points, and in turn, have the highest ranking in this metric.
